##Data cleaning

Changing datatypes

```python
df['Join_Date']=df['Join_Date'].astype('datetime64[ns]')
```

Null values

```python
df.isnull().sum()
```

Filling missing values

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df['Salary']=df['Salary'].fillna(df['Salary'].mean())
```

Checking duplicates

```python
df.duplicated().sum()
```

Removing '-' sign from Phone column

```python
df['Phone']=df['Phone'].astype(str)
```

```python
df['Phone']=df['Phone'].str.replace('-',"",regex=True)
```

```python
df['Phone']=df['Phone'].astype(int)
```

Making new column from Department-Region

```python
df['Department'] = df['Department_Region'].str.split('-').str[0]
df['Region'] = df['Department_Region'].str.split('-').str[1]
```

droping old column

```python
df.drop(columns='Department_Region', inplace=True)
```
